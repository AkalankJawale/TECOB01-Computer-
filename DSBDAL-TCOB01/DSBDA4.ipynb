```python
import pandas as pd
import numpy as np


np.set_printoptions(precision=4, linewidth=100)
mba_salary_df=pd.read_csv('MBA Salary.csv')
mba_salary_df.head(10)

```

```python
mba_salary_df.info()
```

```python
import statsmodels.api as sm
X=sm.add_constant(mba_salary_df['Percentage in Grade 10'])
X.head(5)
```

```python
Y=mba_salary_df['Salary']
from sklearn.model_selection import train_test_split
train_X,test_X,train_Y,test_Y=train_test_split(X,Y,test_size=0.8,random_state=100)
mba_salary_lm=sm.OLS(train_Y,train_X).fit()
print(mba_salary_lm.params)
```

```python
print(mba_salary_df)
```

```python
predictions=mba_salary_lm.predict(test_X)
print(predictions)
```

```python
from sklearn.metrics import mean_squared_error, r2_score  # Import r2_score
test_r2=r2_score(test_Y,predictions)
mse=mean_squared_error(test_Y, predictions)
print(f"R-squared: {test_r2}")
print(f"Mean Squared Error: {mse}")
```

```python
import matplotlib.pyplot as plt
plt.scatter(test_X['Percentage in Grade 10'], test_Y, color='blue', label='Actual values')
plt.plot(test_X['Percentage in Grade 10'], predictions, color='red', label='Fitted line')
plt.xlabel('Percentage in Grade 10')
plt.ylabel('Salary')
plt.title('MBA Salary vs Percentage in Grade 10')
plt.legend()
plt.show()
```
